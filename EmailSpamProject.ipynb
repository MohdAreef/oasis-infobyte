# E-mail spam detection project

```python
import numpy as np
import pandas as pd
```

```python
from collections import Counter
```

# Loading data

```python

df=pd.read_csv("spam.csv",encoding="latin-1")
```

```python
df.head()
```

```python
df.iloc[2].v2
```

# fetching the emails from the data set

```python
emails=df["v2"]
```

```python
emails
```

# splitting each word in emails

```python
words=[]
for i in emails:
    words=words+(i.split(" "))
words
```

# filtering words list that doesn't contain alphabets

```python
for i in range(len(words)):
    if not  words[i].isalpha():
        words[i]=""
```

```python
words_count=Counter(words)
```

```python
len(words_count) 
```

# selecting most frequent 4000 words

```python
word_freq=words_count.most_common(4000)
```

```python
word_freq
```

```python
word_freq.pop(0)
```

```python
word_freq
```

# creating  features of required dataset

```python
features=[]
for i in word_freq:
    features.append(i[0]);
features
```

```python
len(features)      
```

# generating each row with the help of each email

```python
records=[]
for email in emails:
    data=[]
    for word in features:
        data.append(email.split(" ").count(word))
    records.append(data)
```

```python
len(records)
```

```python
label=df.v1[:5572]
```

```python
len(label)
```

```python
X=np.array(records)
Y=np.array(label)
```

```python
X.shape
```

```python
Y=Y.reshape(5572,1)
```

```python
Y.shape
```

```python
X
```

# Let's split the data for training and testing. 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,Y,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier=MultinomialNB()
```

# fitting the  training data to the algorithm

```python
# in train stage naive bayes creates look up table
classifier.fit(X_train,y_train)
```

# testing trained machine learning model

```python
y_predict=classifier.predict(X_test)
```

# Finding accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_predict)
```

# testing my personal emails 

```python
# function which predict wheather email is spam or not.

def search_mail(test_email):
    word_count=[]
    for word in features:
        word_count.append(test_email.split(" ").count(word))
    word_count=np.array(word_count)
    word_count.shape
    y_pre=classifier.predict([word_count])
#     return y_pre;
#     print(y_pre)
    if(y_pre=='spam'):
        print("spam")
    else:
        print("Not spam")
```
